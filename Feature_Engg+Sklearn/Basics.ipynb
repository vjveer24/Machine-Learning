```python
#Understanding the data
data = pd.read_csv('data.csv')
#How big is the dataset?
print(data.shape)
#How does the dataset look like?
print(data.head())
#What are the data types of each column?
print(data.dtypes)
#Are there any missing values in the dataset?
print(data.isnull().sum())
#How many unique values are there in each column?
print(data.nunique())
#How does the data look mathematically?
print(data.describe())
#Are there any duplicate rows in the dataset?
print(data.duplicated().sum())
#How is the correlation between the numerical columns in the dataset?
print(data.corr())
```

```python
#Exploratory Data Analysis (EDA) for univariate analysis
#Univariate analysis is the simplest form of analyzing data. "Uni" means "one", so in other words, your data has only one variable. 
# Univariate analysis doesn’t deal with causes or relationships (unlike regression) and its major purpose is to describe;
# it takes data, summarizes that data and finds patterns in the data.
#Categorical data can be analyzed using frequency tables, bar charts, and pie charts.
import seaborn as sns
#Frequency table for a categorical column
print(data['column_name'].value_counts())
#Bar chart for a categorical column
data['column_name'].value_counts().plot(kind='bar')
plt.show()
#Pie chart for a categorical column
data['column_name'].value_counts().plot(kind='pie')
#Countplot for a categorical column
sns.countplot(x='column_name', data=data)
#Numerical data can be analyzed using histograms, box plots, and density plots.
#Histogram for a numerical column
data['column_name'].plot(kind='hist')
plt.show()
#Box plot for a numerical column
data['column_name'].plot(kind='box')
plt.show()
#Density plot for a numerical column
data['column_name'].plot(kind='density')
plt.show()
#Explaining Different Types of Plots:
#1. Histogram: A histogram is a graphical representation of the distribution of numerical data. 
# It is an estimate of the probability distribution of a continuous variable. 
# It is created by dividing the range of the data into intervals (bins) and counting the number of observations that fall into each bin. 
# The height of each bar represents the frequency of observations in that bin.
#2. Box Plot: A box plot (or whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: 
# minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It can also show outliers in the data. 
# The box represents the interquartile range (IQR) where the middle 50% of the data lies, 
# and the whiskers extend to the minimum and maximum values within 1.5 * IQR from Q1 and Q3.
#3. Density Plot: A density plot is a smoothed, continuous version of a histogram estimated from the data. 
# It shows the distribution of a numerical variable and is useful for visualizing the probability density function of the variable. 
# It is created using kernel density estimation (KDE) and provides a clearer view of the distribution shape compared to histograms, 
# especially for smaller datasets.
```

```python
#EDA for bivariate analysis and multivariate analysis
#Scatter plot for two numerical columns
data.plot(kind='scatter', x='column_name1', y='column_name2',hue='column_name3')
#Bar plot for a categorical column and a numerical column
data.groupby('categorical_column')['numerical_column'].mean().plot(kind='bar')
#Box plot for a categorical column and a numerical column
data.boxplot(column='numerical_column', by='categorical_column')  
#Distplot for a categorical column and a numerical column
sns.distplot(data[data['categorical_column']=='category1']['numerical_column'], label='category1')
sns.distplot(data[data['categorical_column']=='category2']['numerical_column'], label='category2')
plt.legend()
plt.show()
#Heatmap for correlation between numerical columns
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
#Cluster map for correlation between numerical columns
sns.clustermap(data.corr(), annot=True, cmap='coolwarm')
#Pair plot for numerical columns
sns.pairplot(data)
#Line plot for a numerical column
data['numerical_column'].plot(kind='line')
```
