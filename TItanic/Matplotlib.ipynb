### Matplotlib

```python
#Basics
import matplotlib.pyplot as plt
import numpy as np

x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])

plt.plot(x,y)
plt.show()
```

```python
# bar chart = compare categories of data by representing each category with a bar

categories = np.array(["Grains", "Fruits", "Vegetables", "Protein", "Dairy", "Sweets"])
values = np.array([4,3,2,5,3,1])

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.bar(categories, values, color="skyblue")
#plt.barh(categories, values) horizontal bars

plt.show()
```

```python
#grid() = helps make plots easier to read by adding reference lines

x=[1,2,3,4,5]
y=[5,10,15,20,25]

#axis could be x, y or both
plt.grid(axis="y", linewidth=2, color="grey",
            linestyle="dashed")

plt.plot(x,y)
plt.show()
```

```python
# Histogram = A visual representation of the distribution of quantitative data.
#             they group values into bins (intervals)
#             and counts how many fall in each range

#loc=mean, scale=std deviation
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10, color="lightgreen", edgecolor="black")

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("# of Students")

plt.show()
```

```python
# adding labels 
x=np.array([2023,2024,2025,2026])
y1=np.array([15,25,30,20])
y2=np.array([17,23,38,5])
y3=np.array([14,15,20,30])

plt.title("Class size", fontsize=20, family="Arial",
            fontweight="bold",color="blue")

plt.xlabel("Year",fontsize=20,family="Arial",
            fontweight="bold",color="grey")

plt.ylabel("Students",fontsize=20,family="Arial",
            fontweight="bold",color="grey")

plt.tick_params(axis="both",colors="grey")

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.xticks(x)

plt.show()
```

```python
# Pie chart = Circular chart divided into slices to show percentages of the total.
#             Good for visualizing distribution among categories

categories = np.array(["Freshmen", "Sophomores", "Juniors", "Seniors"])
values = np.array([1080,705,457,375])
colors=['#1f77b4','#aec7e8','#ff7f0e','#ffbb78']

plt.pie(values, labels=categories, autopct="%1.1f%%", colors=colors,
        explode=[0,0,0,0.2], shadow=True, startangle=180)
plt.title(">_<")
plt.show()
```

```python
#customizing plot
x=np.array([1,2,3,4])
y1=np.array([0,1,0,1])
y2=np.array([4,3,2,1])
#ms=marker size, mfc=markerfacecolor, mec=markeredgecolor
#linestyle can be dotted, solid, dashed, dashdot, none

line_style = dict(marker=".", ms=20,
        mfc="cyan", mec="cyan",
        linestyle="solid", linewidth=2)

plt.plot(x,y1,**line_style,color="red")
plt.plot(x,y2, **line_style, color="green")
plt.show()
```

```python
# Scatter Graphs: Shows the relationship between two variables 
#                 Helps to identify a correlation (+, -, None) 
#                 Example: Study Hours vs. Test Scores

x1 = np.array([0,1,1,2,3,4,5,6,7,7,8]) # Hours studied
y1 = np.array([55,60,65,62,68,70,75,78,82,85,87]) #Grades

x2 = np.array([0,1,2,2,3,4,5,6,7,8,8]) # Hours studied
y2 = np.array([50,58,65,70,72,78,83,88,92,95,97]) #Grades

plt.scatter(x1,y1, color="skyblue", alpha = 0.5, s=100, label="Class A")
plt.scatter(x2,y2, color="red", alpha = 0.5, s=100, label="Class B")


plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.legend()
plt.show()
```

```python
#Making subplots
# figure = the entire canvas
# ax = a single plot (subplot)

x = np.array([1,2,3,4,5])

figure, axes = plt.subplots(2,2)

axes[0,0].plot(x,x*2, color="red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x,x**2, color="blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x,x**3, color="green")
axes[1,0].set_title("x**3")

axes[1,1].plot(x,x**4, color="purple")
axes[1,1].set_title("x**4")

#to avoid overlapping
plt.tight_layout()
plt.show()
```

```python
#using pandas and matplotlib together
import pandas as pd
df = pd.read_csv("pokemondataset.csv")

type_count = df["Type 1"].value_counts(ascending=True)

plt.barh(type_count.index, type_count.values, color="cyan", 
            edgecolor="black")
plt.title("# of Pokemon by Primary type")
plt.xlabel("Count")
plt.ylabel("Type")
plt.tight_layout()
plt.show()
```
