```python

data = [100,102,104]
#index are optional to set
#default indexes are 0,1,2....
series = pd.Series(data, index=["a","b","c"])
print(series)
#locating and accessing using index
print(series.loc["b"])
#using integer index
print(series.iloc[1])
print(series[series>=102])

calories={"day 1":1750,"day 2":2100, "day 3":1700}
s=pd.Series(calories)
s.loc["day 3"] += 500
print(s)
```

```python
#Data Cleaning = the process of fixing/removing: 
#                incomplete,incorrect, or irrelevant dat. 
#                ~75% of work done with pandas is data cleaning
df=pd.read_csv("pokemondataset.csv")

#1. Drop irrelevant columns
df = df.drop(columns=["Legendary", "Generation","Sp. Atk", "Sp. Def","#"])

#2. Handle missing data
df = df.dropna(subset=["Type 2"])
# to replace these values instead
# df = df.fillna({"Type 2": "None"})

#3. Fix inconsistent values
df["Type 1"] = df["Type 1"].replace({"Grass": "GRASS", "Fire": "FIRE"})

#4. Standardize texts
df["Name"] = df["Name"].str.lower()

#5. Fix data types
df["Attack"] = df["Attack"].astype(float)

#6. Remove duplicate values
# df = df.drop_duplicates()

print(df)
```
